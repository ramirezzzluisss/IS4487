## Clean up the data
Did you notice anything unusual about the "house" amounts? 

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Slice the data  
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2) 
```

```python
# two steps to clean data

#cleaned by customers who left 
df_leave = df[(df['leave'] == 'LEAVE')]

#cleaned by customers who stayed
df_stay = df[(df['leave'] == 'STAY')]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
# count plot of considerations for customers that left 
sns.countplot(x ='considering_change_of_plan', data = df_leave)
 
# Show the plot
plt.show()
```

```python
# count plot of satisfaction level of customers that stayed
sns.countplot(x ='reported_satisfaction', data = df_stay)
 
# shot the plot
plt.show()
```
